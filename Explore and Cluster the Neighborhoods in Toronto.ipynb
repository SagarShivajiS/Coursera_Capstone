Download and create list of postal codes of canada table from wikipedia

```python
WIKI_URL = ('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
df = pd.read_html(WIKI_URL, header=0) # set header with row #0

df_Canada_PostalCodes = df[0]
```

Keeping only boroughs having assigned neighborhoods

```python
df_Canada_PostalCodes = df_Canada_PostalCodes[df_Canada_PostalCodes.Borough != 'Not assigned']
```

Grouping

```python
df_Canada_PostalCodes = df_Canada_PostalCodes.groupby(by=['Postal code','Borough']).agg(lambda x: ','.join(x))

df_Canada_PostalCodes.reset_index(level=['Postal code','Borough'], inplace=True)

df_Canada_PostalCodes
```

Exploring and clustering the neighborhoods in Toronto

```python
print('The dataframe has {} Borough and {} Neighbourhood.'.format(
        len(df_Postal_Geo['Borough'].unique()),
        df_Postal_Geo.shape[0]
    )
)
```

```python
Toronto=df_Postal_Geo[df_Postal_Geo['Borough'].str.contains('Toronto')]
Toronto
```

```python
address = 'Toronto, TR'

geolocator = Nominatim(user_agent="tr_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

## Ans 3)

```python
Toronto_map = folium.Map(location=[latitude, longitude], zoom_start=10)

for lat, lng, borough, neighborhood in zip(df_Postal_Geo['Latitude'], df_Postal_Geo['Longitude'], 
                                           df_Postal_Geo['Borough'], df_Postal_Geo['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(Toronto_map)  
    
Toronto_map
```
